```
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
import sklearn
```

```
#Wczytanie danych
df2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Nauka/PCA/wine.data', header=None)
df=df2.copy()
df.head()
```

```
dane = df.iloc[:,1:]
target= df.iloc[:,0]
```

```
dane[:3]
```

```
target[:3]
```

```
target.value_counts()
```

```
#Podział na zbiór treningowy i testowy
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(dane, target)
print("Ilość danych train:", X_train.shape[0])
print("Ilość danych test:",X_test.shape[0])
```

```
# Standaryzacja danych treingowych
from sklearn.preprocessing import StandardScaler
```

```
scaler = StandardScaler()
X_train_s=scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```
X_train_s[:3]
```

```
# PCA
from sklearn.decomposition import PCA
```

```
pca=PCA(n_components=3)
X_train_p=pca.fit_transform(X_train_s)
X_test_p = pca.transform(X_test_s)
```

```
X_train_p[:3]
```

```
df_pca=pd.DataFrame(data=np.c_[X_train_p, y_train], columns=[1,2,3,'target'])
df_pca.head()
```

```
px.scatter_3d(df_pca,
           x=1,
           y=2,
           z=3,
           color='target', width= 700, height=500)
```

```
# wyjaśniona wariancja
a= pca.explained_variance_
a
```

```
# wartości skumulowane 
b=(pca.explained_variance_).cumsum()
b
```

```
df_var=pd.DataFrame(data=[a,b])
df_var=df_var.T
df_var.columns=['wyjaśniona wariancja','skumulowane']
df_var['index']=df_var.index+1
df_var
```

```
fig= go.Figure(data=[go.Bar(x=df_var['index'], y=df_var['wyjaśniona wariancja']),
                     go.Scatter(x=df_var['index'], y=df_var['skumulowane'])])
fig.show()
                            
```

```
# wyjaśnione zostało 79% wariancji
```

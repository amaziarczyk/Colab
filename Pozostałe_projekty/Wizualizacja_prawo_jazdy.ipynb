```
import pandas as pd
```

```
df = pd.read_csv("/content/drive/My Drive/Colab Notebooks/prawo_jazdy/nowe_prawa_jazdy_wg_plci_i_wieku(1-5.2020).csv", encoding = 'windows-1250', sep="|")
```

```
df.head()
```

```
# PRZYGOTOWANIE DANYCH

#sprawdzenie liczności grupy wojewodztwo
df['WOJEWODZTWO'].value_counts()
```

```
#grupowanie według wojewodztwa
df_woj = df.groupby(['WOJEWODZTWO'])['LICZBA'].sum().reset_index()
df_woj
```

```
#sortowanie według liczby
df_sort = df_woj.sort_values(by="LICZBA", ascending= False)
df_sort
```

```
# WYKRESY w Plotly

import plotly.express as px
```

```
#W jakim województwie wydano najmniej praw jazdy w okresie 1-5.2020 roku?

wykres = px.bar(df_sort,
                 x = "WOJEWODZTWO",
                 y = "LICZBA",
                 title = "Prawa jazdy w 2020",
                 color = "WOJEWODZTWO",
                )
wykres.show()
```

```
#W jakim miesiącu wydano najwięcej praw jazdy kobietom wg. wszystkich województw?

# PRZYGOTOWANIE DANYCH
# zmiana formatu daty i stworzenie kolumny miesiąc
df['DATA_MC']=pd.to_datetime(df["DATA_MC"])
df['month']=df["DATA_MC"].dt.month
df.head()
```

```

#filtrowanie kobiet
df_kobiety = df[df['PLEC']=='K']
df_kobiety.head(10)
```

```
# grupowanie wg daty
df_group= df_kobiety.groupby('month')['LICZBA'].sum().reset_index()
df_group
```

```
# WYKRES 
wykres_2 = px.bar(df_group,
            x = "month",
            y = "LICZBA",
            title = "Kobiety- prawo jazdy",
            color="month")
wykres_2.show()
```
